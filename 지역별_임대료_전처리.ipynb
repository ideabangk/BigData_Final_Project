```python
import pandas as pd
df = pd.read_csv('한국부동산원_상업용부동산 임대동향조사_임대정보_분기별 층별임대료 및 층별효용비율(중대형상가)_20240331.csv', encoding='ANSI')
df.info()
```

# 서울 필터링

```python
df['시도 광역상권 하위상권 구분'].unique()
```

```python
# 서울 임대료 데이터만 필터링
df = df[(df['시도 광역상권 하위상권 구분'].str.startswith('서울')) & (df['시도 광역상권 하위상권 구분'].str.endswith('임대료'))].reset_index(drop=True)
df
```

```python
df['시도 광역상권 하위상권 구분'].unique()
```

# 컬럼 구분

```python
# 데이터 분리: 정규식을 사용하여 '시도', '광역상권', '하위상권'으로 나누기
df[['시도', '광역상권', '하위상권']] = df['시도 광역상권 하위상권 구분'].str.extract(
    r'(?P<시도>서울)\s*(?P<광역상권>[^\s]*?)?\s*(?P<하위상권>[^\s]+임대료)'
)

# '임대료' 제거 및 공백 처리
df['하위상권'] = df['하위상권'].str.replace('임대료', '', regex=False).str.strip()


df = df.drop(columns=['시도 광역상권 하위상권 구분'])

df
```

# 결측값 채우기

```python
df.isna().sum()
```

```python
df[df['시도'].isnull()]
```

```python
# 정규표현식에 맞지않아 추출되지 못한 NaN 값 채우기
df.loc[df['광역상권'].isna(), '시도'] = '서울'
df.loc[1,'광역상권'] = '도심'
df.loc[10,'광역상권'] = '강남'
df.loc[25,'광역상권'] = '영등포신촌'
df.loc[33,'광역상권'] = '기타'

# 임대료 데이터 NaN값 평균값으로 대체
df['2022_1분기_6-10층'] = df['2022_1분기_6-10층'].fillna(df['2022_1분기_6-10층'].mean())
df['2022_2분기_6-10층'] = df['2022_2분기_6-10층'].fillna(df['2022_2분기_6-10층'].mean())
df['2022_3분기_6-10층'] = df['2022_3분기_6-10층'].fillna(df['2022_3분기_6-10층'].mean())
df['2022_4분기_6-10층'] = df['2022_4분기_6-10층'].fillna(df['2022_4분기_6-10층'].mean())
df['2023_1분기_6-10층'] = df['2023_1분기_6-10층'].fillna(df['2023_1분기_6-10층'].mean())
df['2023_2분기_6-10층'] = df['2023_2분기_6-10층'].fillna(df['2023_2분기_6-10층'].mean())
df['2023_3분기_6-10층'] = df['2023_3분기_6-10층'].fillna(df['2023_3분기_6-10층'].mean())
df['2023_4분기_6-10층'] = df['2023_4분기_6-10층'].fillna(df['2023_4분기_6-10층'].mean())
df['2024_1분기_6-10층'] = df['2024_1분기_6-10층'].fillna(df['2024_1분기_6-10층'].mean())
# 광역상권, 하위상권 NaN값 공백으로 채우기
df[['광역상권','하위상권']] = df[['광역상권','하위상권']].fillna('')

df
```

# 분석에 용이한 형태로 데이터 구조 변환

```python
# 데이터 구조 변환
melt_df = df.melt(id_vars=['시도', '광역상권', '하위상권'], var_name='연도_기간_층', value_name='임대료')

# '기간_층' 열을 '기간'과 '층'으로 분리
melt_df[['연도', '기간', '층']] = melt_df['연도_기간_층'].str.split('_', expand=True)
melt_df = melt_df.drop(columns=['연도_기간_층'])  # 기존 '기간_층' 열 삭제

# 임대료 데이터는 단위가 천원/m^2 이기 때문에 평당 임대료로 전환
melt_df['임대료'] = melt_df['임대료'] * 1000 * 3.305785
melt_df['임대료'] = melt_df['임대료'].astype('int64')
melt_df = melt_df.rename(columns={'임대료':'평당_임대료'})

melt_df
```

```python
melt_df.to_csv('상권별_상업용_중대형상가_임대료.csv', encoding='utf-8' ,index=False)
```
